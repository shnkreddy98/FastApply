```python
from selenium import webdriver
from html2text import html2text
from bs4 import BeautifulSoup
import json

import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
def getContent(url):
    options = webdriver.ChromeOptions()
    options.add_argument("--headless=new")
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    html_content = driver.page_source
    driver.quit()
    return html_content

def getCompanyName(text, url):
    url_arr = url.split("/")
    if "myworkdayjobs" in url_arr[2] or "zohorecruit" in url_arr[2]:
        company = url_arr[2].split(".")[0]
    elif "jobs.ashbyhq" in url_arr[2] or \
         "job-boards.greenhouse" in url_arr[2] or \
         "jobs.lever" in url_arr[2] or \
         "ats.rippling" in url_arr[2]:
        company = url_arr[3]
    else:
        doc = nlp(text)
        items = [ent.text for ent in doc.ents if ent.label_=='ORG']
        company = Counter(items).most_common(1)[0][0]
    
    try:
        clean_name = company.split("\n")
        return clean_name
    except:
        return company
    
```

```python
import time

for url in urls:
    html_content = getContent(url)

    text_content = html2text(html_content)

    try:
        if len(text_content)>10:
            company = getCompanyName(text_content, url)
        else:
            soup = BeautifulSoup(html_content, features="lxml")

            data = json.loads(soup.find('script', type='application/ld+json').string)
            company = getCompanyName(str(data['description']), url)
    except:
        company = "NotFound"
    print(company, url)
    time.sleep(1)
```
