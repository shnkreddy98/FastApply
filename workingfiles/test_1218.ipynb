```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
import os
from dotenv import load_dotenv
import time

load_dotenv()

email_password = os.getenv("ssr")

def send_emails(from_email, possible_emails, subject, body, resumefilepath)
    fromaddr = from_email
    toaddrs = possible_emails

    for toaddr in toaddrs:
        time.sleep(0.5)
        msg = MIMEMultipart()
        msg['From'] = fromaddr
        msg['To'] = toaddr
        msg['Subject'] = subject

        body = body
        msg.attach(MIMEText(body, 'plain'))

        filename = os.path.basename(resumefilepath)
        file_path = resumefilepath

        with open(file_path, "rb") as attachment:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header('Content-Disposition', f"attachment; filename={filename}")
            msg.attach(part)

        try:
            server = smtplib.SMTP('smtp.gmail.com', 587)
            server.starttls()
            server.login(fromaddr, email_password)
            text = msg.as_string()
            server.sendmail(fromaddr, toaddr, text)
            print(f"Email sent successfully to {toaddr}")
        except Exception as e:
            print(f"Failed to send email to {toaddr}. Error: {e}")
        finally:
            server.quit()
```

```python
import socket
import socks # PySocks

from smtplib import SMTP

class SocksSMTP(SMTP):

    def __init__(self,
            host='',
            port=0,
            local_hostname=None,
            timeout=socket._GLOBAL_DEFAULT_TIMEOUT,
            source_address=None,
            proxy_type=None,
            proxy_addr=None,
            proxy_port=None,
            proxy_rdns=True,
            proxy_username=None,
            proxy_password=None,
            socket_options=None):

        self.proxy_type=proxy_type
        self.proxy_addr=proxy_addr
        self.proxy_port=proxy_port
        self.proxy_rdns=proxy_rdns
        self.proxy_username=proxy_username
        self.proxy_password=proxy_password
        self.socket_options=socket_options
        # if proxy_type is provided then change the socket to socksocket
        # else behave like a normal SMTP class.
        if self.proxy_type:
            self._get_socket = self.socks_get_socket

        super(SocksSMTP, self).__init__(host, port, local_hostname, timeout, source_address)

    def socks_get_socket(self, host, port, timeout):
        if self.debuglevel>0:
            self._print_debug('connect: to', (host, port), self.source_address)
        return socks.create_connection((host, port),
                timeout=timeout,
                source_address=self.source_address,
                proxy_type=self.proxy_type,
                proxy_addr=self.proxy_addr,
                proxy_port=self.proxy_port,
                proxy_rdns=self.proxy_rdns,
                proxy_username=self.proxy_username,
                proxy_password=self.proxy_password,
                socket_options=self.socket_options)
```

```python
import socket
import socks
import smtplib
from smtplib import SMTP

class SocksSMTP(SMTP):
    def __init__(self,
                 host='',
                 port=0,
                 proxy_host=None,
                 proxy_port=None,
                 local_hostname=None,
                 timeout=socket._GLOBAL_DEFAULT_TIMEOUT,
                 source_address=None):
        self.proxy_host = proxy_host
        self.proxy_port = proxy_port

        # If proxy_host and proxy_port are provided, use a SOCKS proxy socket
        if self.proxy_host and self.proxy_port:
            self._get_socket = self.socks_get_socket

        super().__init__(host, port, local_hostname, timeout, source_address)

    def socks_get_socket(self, host, port, timeout):
        if self.debuglevel > 0:
            self._print_debug('connect: to', (host, port), self.source_address)
        return socks.create_connection(
            (host, port),
            timeout=timeout,
            source_address=self.source_address,
            proxy_type=socks.SOCKS5,
            proxy_addr=self.proxy_host,
            proxy_port=self.proxy_port
        )

def verify_email(to_addr, from_addr, smtp_host, smtp_port, proxy_host, proxy_port):
    try:
        with SocksSMTP(
            host=smtp_host,
            port=smtp_port,
            proxy_host=proxy_host,
            proxy_port=proxy_port,
            timeout=10
        ) as smtp:
            smtp.starttls()  # Important for Gmail SMTP
            smtp.helo()
            smtp.mail(from_addr)
            resp = smtp.rcpt(to_addr)
            
            if resp[0] == 250:
                return True, "Email address is valid"
            elif resp[0] == 550:
                return False, "Email address is invalid"
            else:
                return False, f"Unexpected response code: {resp[0]}"
                
    except smtplib.SMTPServerDisconnected as err:
        return False, f"SMTP connection error: {str(err)}"
    except socket.gaierror as err:
        return False, f"DNS resolution error: {str(err)}"
    except Exception as err:
        return False, f"Verification failed: {str(err)}"

# Usage example
if __name__ == "__main__":
    to_addr = "uncancceledpotato@gmail.com"
    from_addr = "sreddy2095@gmail.com"
    smtp_host = "gmail-smtp-in.l.google.com"
    smtp_port = 587
    proxy_host = "13.36.113.81"
    proxy_port = 3128

    is_valid, message = verify_email(
        to_addr=to_addr,
        from_addr=from_addr,
        smtp_host=smtp_host,
        smtp_port=smtp_port,
        proxy_host=proxy_host,
        proxy_port=proxy_port
    )
    
    print(f"Result: {is_valid}")
    print(f"Message: {message}")
```

```python
import threading
import queue
import requests

q = queue.Queue()
valid_proxies = []

with open("../.proxies_list.txt", "r") as f:
    proxies = f.read().split("\n")
    for p in proxies:
        q.put(p)


def check_proxies():
    global q
    while not q.empty():
        proxy = q.get()
        try:
            res = requests.get("http://ipinfo.io/json", 
                               proxies={"http": proxy,
                                        "https:": proxy})
            
        except:
            continue

        print(res.status_code)
        if res.status_code==200:
            valid_proxies.append(proxy)


for _ in range(10):
    threading.Thread(target=check_proxies).start()

```
